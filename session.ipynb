```python
import os
import sys

# Disable problematic Hadoop filesystem providers
os.environ['HADOOP_CONF_DIR'] = '/tmp/hadoop-conf'
os.environ['HADOOP_HOME'] = '/tmp'

# Add configuration for Java security
os.environ['_JAVA_OPTIONS'] = '-Dcom.sun.jndi.rmi.object.trustURLCodebase=false -Dcom.sun.jndi.cosnaming.object.trustURLCodebase=false'

# Prevent the security manager from initializing
import findspark
findspark.init()

from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .master("local[*]") \
    .appName("test") \
    .config("spark.driver.host", "127.0.0.1") \
    .config("spark.hadoop.fs.viewfs.enable", "false") \
    .config("spark.hadoop.javax.security.auth.useSubjectCredsOnly", "false") \
    .config("spark.hadoop.security.authorization", "false") \
    .config("fs.file.impl", "org.apache.hadoop.fs.LocalFileSystem") \
    .config("fs.AbstractFileSystem.file.impl", "org.apache.hadoop.fs.local.LocalFileSystem") \
    .config("spark.hadoop.user.name", "spark") \
    .config("spark.sql.warehouse.dir", "file:///tmp/spark-warehouse") \
    .config("spark.shuffle.service.enabled", "false") \
    .config("spark.dynamic.allocation.enabled", "false") \
    .config("spark.serializer", "org.apache.spark.serializer.JavaSerializer") \
    .getOrCreate()

print("Spark version:", spark.version)
print("Spark session created successfully!")
```
